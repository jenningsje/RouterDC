```python
import json
import os
from matplotlib import pyplot as plt


with open("/data/home/chensh/projects/LLM_router/logs/router_debug/split2_model5_2_cluster_3/slw_0_clw_1_clw_2_0_cos_tk_1_lk_3_lr_5e-5_step_1000_t_1_seed_0/training_log.json", 'r') as f:
    training_result = json.load(f)

step = 50
print(training_result[1].keys())
# .keys()

dataset_prefix1 = "./datasets/split2_model5_2_cluster_3/"
dataset_prefix2 = "./datasets/split2_model5_2/"

# dataset_prefix1 = "./datasets/split2/"
# dataset_prefix2 = "./datasets/split2/"


training_result[1]
```

```python
training_result_name = dataset_prefix1 + "gsm8k-train.json"
training_result_name2 = dataset_prefix2 + "gsm8k-test.json"

plot_x = [result[training_result_name][1] for result in training_result]

plt.plot([step *i for i in range(len(plot_x))], plot_x)

plot_x = [result[training_result_name2][1] for result in training_result ]
plt.xlabel(f"{training_result_name}+{training_result_name2}")

plt.plot([step *i for i in range(len(plot_x))], plot_x)
plt.legend([f"{training_result_name} acc", f"{training_result_name2} acc"])

plt.plot([step *i for i in range(len(plot_x))], [0.6963 for i in range(len(plot_x))])
```
