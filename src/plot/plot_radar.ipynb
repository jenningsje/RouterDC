```python
import numpy as np
import matplotlib.pyplot as plt
from math import pi

# 定义每个维度的标签
categories = ['A', 'B', 'C', 'D', 'E', 'F']

# 假设我们有三组数据
values1 = [20, 35, 30, 35, 27, 25]
values2 = [25, 32, 34, 20, 25, 30]
values3 = [30, 22, 29, 35, 33, 35]

# 计算每个维度的角度
N = len(categories)
angles = [n / float(N) * 2 * pi for n in range(N)]
values1 += values1[:1]
values2 += values2[:1]
values3 += values3[:1]
angles += angles[:1]

# 初始化雷达图
ax = plt.subplot(111, polar=True)

# 绘制第一组数据
ax.plot(angles, values1, linewidth=1, linestyle='solid', label='Group A')
ax.fill(angles, values1, 'b', alpha=0.1)

# 绘制第二组数据
ax.plot(angles, values2, linewidth=1, linestyle='solid', label='Group B')
ax.fill(angles, values2, 'r', alpha=0.1)

# 绘制第三组数据
ax.plot(angles, values3, linewidth=1, linestyle='solid', label='Group C')
ax.fill(angles, values3, 'g', alpha=0.1)

# 添加每个维度的标签
plt.xticks(angles[:-1], categories)

# 设置y轴的范围和标签
ax.set_rscale('linear')
ax.set_rticks([10, 20, 30, 40, 50, 60, 70], [])
ax.set_rlabel_position(0)

# 添加图例
plt.legend(loc='upper right', bbox_to_anchor=(0.1, 0.1))

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from math import pi

# 定义每个维度的标签
categories = ['A', 'B', 'C', 'D', 'E', 'F']

# 假设我们有三组数据
values1 = [20, 35, 30, 35, 27, 25]
values2 = [25, 32, 34, 20, 25, 30]
values3 = [30, 22, 29, 35, 33, 35]

# 计算每个维度的角度
N = len(categories)
angles = [n / float(N) * 2 * pi for n in range(N)]
values1 += values1[:1]
values2 += values2[:1]
values3 += values3[:1]
angles += angles[:1]

# 初始化雷达图
fig, ax = plt.subplots(figsize=(6, 6), subplot_kw=dict(polar=True))

# 绘制多边形的网格线
ax.set_theta_offset(pi / 2)  # 旋转起始位置
ax.set_theta_direction(-1)   # 逆时针方向

# 绘制多边形的外边框
ax.set_thetagrids(np.degrees(angles[:-1]), labels=categories)  # 注意这里使用 angles[:-1]

# 设置雷达图的网格线为多边形
for i in range(N):
    angle_rad = angles[i]
    ax.plot([angle_rad, angle_rad], [0, 70], color='gray', linestyle='solid', linewidth=0.5)

# 绘制数据
ax.plot(angles, values1, 'b-', linewidth=2, label='Group A')
ax.fill(angles, values1, color='blue', alpha=0.25)
ax.plot(angles, values2, 'r-', linewidth=2, label='Group B')
ax.fill(angles, values2, color='red', alpha=0.25)
ax.plot(angles, values3, 'g-', linewidth=2, label='Group C')
ax.fill(angles, values3, color='green', alpha=0.25)

# 添加图例
plt.legend(loc='upper right', bbox_to_anchor=(0.1, 0.1))

plt.show()
```
