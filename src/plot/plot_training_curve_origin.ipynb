```python
import json
import os
from matplotlib import pyplot as plt


with open("/data/home/chensh/projects/LLM_router/logs/router_debug/origin/top_k_2_last_k_3_learning_rate_0.00005_step_1000/training_log.json", 'r') as f:
    training_result = json.load(f)

step = 20
print(training_result[1].keys())
# .keys()

dataset_prefix = "./datasets/"
```

```python
training_result_name = dataset_prefix + "gsm8k-train.json"
training_result_name2 = dataset_prefix + "gsm8k-test.json"

plot_x = [result[training_result_name][1] for result in training_result]

plt.plot([step *i for i in range(len(plot_x))], plot_x)

plot_x = [result[training_result_name2][1] for result in training_result ]
plt.xlabel(f"{training_result_name}+{training_result_name2}")

plt.plot([step *i for i in range(len(plot_x))], plot_x)
plt.legend([f"{training_result_name} acc", f"{training_result_name2} acc"])

plt.plot([step *i for i in range(len(plot_x))], [0.7125 for i in range(len(plot_x))])

plot_x = [result[training_result_name][0] for result in training_result]

plt.plot([step *i for i in range(len(plot_x))], plot_x)

plot_x = [result[training_result_name2][0] for result in training_result ]
plt.plot([step *i for i in range(len(plot_x))], plot_x)
```
